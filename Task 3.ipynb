## StateWise most Sales

```python
state_sale = df.groupby(by = 'State').sum()['Sales'].sort_values(ascending = False).reset_index()
state_sale
```

```python

plt.figure(figsize=(15,12))
sb.barplot(x = 'State', y = 'Sales',data = state_sale)
plt.xticks(rotation = 90)
plt.show()
```

### Sales are maximum in California, New York, Texas,(in decreasing order respectively) and
### Sales are minimum in Wyoming, West Virginia, North Dakota(in increasing order respectively)

## StateWise most Sales

```python
state_prof = df.groupby(by = 'State').sum()['Profit'].sort_values(ascending = False).reset_index()
state_prof
```

```python

plt.figure(figsize=(15,12))
sb.barplot(x = 'State', y = 'Profit',data = state_prof)
plt.xticks(rotation = 90)
plt.show()
```

### Profit is maximum at California, NewYork,Washington respectively.
### Profit is minimum at Texas, Chio, Pennsylvania respectively
